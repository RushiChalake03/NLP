#### Stemming is a text preprocessing technique that reduces words to their root or stem form by removing prefixes or suffixes.

Notice that the stemmed word may not always be a valid dictionary word (e.g., studies → studi).

Why is it used?

    -Reduces vocabulary size.
    -Treats different forms of a word as the same term.
    -Improves efficiency in NLP tasks such as text classification and information retrieval.

##### Why is stemming less accurate than lemmatization?

Answer: Because stemming simply removes prefixes/suffixes using predefined rules and does not consider the word's actual meaning or context, which can result in non-dictionary words.

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
print(stemmer.stem('Playing'))
```

```python
print(stemmer.stem('Studying'))
```

```python
sentence = "The boys are playing the cricket and girls are studying"

print(stemmer.stem(sentence))
```

#### RegexpStemmer Class

NLTK has RegexpStemmer Class which can help us for stemming using regular expression. It simply takes one regular expression and removes the prefix or suffix that matches with regular expression.

```python
from nltk.stem import RegexpStemmer
```

```python
stemming = RegexpStemmer('ing$|s$|e$|able$')
```

```python
stemming.stem('typing')
```

```python
stemming.stem('grass')
```

```python
stemming.stem('ingtyping')
```

#### SnowballStemmer

This is better than PorterStemmer


```python
from nltk.stem import SnowballStemmer
```

```python
snowball_stemmer = SnowballStemmer('english')
```

```python
snowball_stemmer.stem(sentence)
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(sentence)
```

```python
for word in words:
    print(word+"----->"+snowball_stemmer.stem(word))
```
