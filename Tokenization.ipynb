##### Tokenization is the process of breaking a text document into smaller units called tokens, such as words, subwords, characters, or sentences.

Why is it used?

Machine learning and NLP models cannot directly understand raw text. Tokenization converts text into smaller meaningful units that can be processed and transformed into numerical representations.

Interview Answer (One-Liner)

Tokenization is the process of splitting text into smaller units called tokens (words, subwords, characters, or sentences) so that NLP models can process and analyze the text effectively.

Types of Tokenization:
1. Word Tokenization
"I love NLP" → ["I", "love", "NLP"]
2. Sentence Tokenization
"Hello. How are you?" → ["Hello.", "How are you?"]
3. Subword Tokenization (used in BERT, GPT)
"unhappiness" → ["un", "happi", "ness"]
4. Character Tokenization
"NLP" → ["N", "L", "P"]

For interviews, remember:

Text → Tokenization → Numerical Representation (BoW/TF-IDF/Embeddings) → Model Training.

```python
corpus = """Welcome to NLP's world. You will have fun. 
Enjoy the journey! All the best."""
```

```python
print(corpus)
```

#### sent_tokenize

```python
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
documents
```

#### word_tokenize

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

#### wordpunct_tokenize

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

#### TreebankWordTokenizer

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokens = tokenizer.tokenize(corpus)
```

```python
tokens
```
