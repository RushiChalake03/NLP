Stopwords are frequently occurring words in a language that are often removed during text preprocessing because they carry little semantic meaning and may not contribute significantly to NLP tasks such as text classification, sentiment analysis, or information retrieval.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
stop_words
```

```python
paragraph = """A. P. J. Abdul Kalam often encouraged young people to dream big and work hard to achieve their goals.
In one of his famous messages to students, he emphasized that dreams are not what we see while sleeping,
but dreams are those that do not let us sleep. He believed that education, determination, and continuous
learning are the keys to personal and national development. Dr. Kalam inspired millions of people by
urging them to develop a scientific mindset, embrace innovation, and contribute positively to society. 
His speeches continue to motivate students and professionals to pursue excellence and serve the nation with dedication."""
```

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(paragraph)
```

```python
sentences
```

```python
from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word, pos = 'v') for word in words if word not in set(stop_words)]
    sentences[i] = ' '.join(words)
```
